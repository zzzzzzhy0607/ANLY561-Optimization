### Question2
###### a)
The constraints are:
$$h_1=x_1-1\leqslant0,  h_2=-x_1\leqslant0$$
$$h_3=x_2-1\leqslant0,  h_4=-x_2\leqslant0$$
$$h_5=x_3-1\leqslant0,  h_6=-x_3\leqslant0$$
