### Question 1

Let $f({\bf x})=||{\bf y}-{\bf x}||^2_2$ and $g({\bf x})={\bf v^Tx}-b$ 

Then we have  $\nabla f({\bf x})=2({\bf x}-{\bf y})$ and $\nabla g({\bf x})={\bf v}$

Since ${\bf v}$ is nonzero all the time, there is always a $\lambda$ such that $\nabla f({\bf x})=\lambda\nabla g({\bf x})$

Then we have $2({\bf x}-{\bf y})=\lambda {\bf v}$ and ${\bf x}=\frac{\lambda {\bf v}}{2}+{\bf y}$

Plug ${\bf x}$ into ${\bf v^T x}-b=0$, we get ${\bf v^T}(\frac{\lambda {\bf v}}{2}+{\bf y})-b=0$, therefore $\lambda=\frac{2}{||{\bf v}||^2_2}(b-{\bf v^Ty})$

${\bf x}=\frac{\lambda {\bf v}}{2}+{\bf y}=\frac{\frac{2{\bf v}}{||{\bf v}||^2_2}(b-{\bf v^Ty}) }{2}+{\bf y}=\frac{{\bf v}}{||{\bf v}||^2_2}(b-{\bf v^Ty})+{\bf y}$<br><br>  

${\bf x^*}=\frac{{\bf v}}{||{\bf v}||^2_2}(b-{\bf v^Ty})+{\bf y}$<br><br><br>

###### e)

$$\begin{align}
\underset{u \in R^d}{min}||X-X{\bf uu^T}||^2_{Fro}&=\underset{u \in R^d}{min}trace((X-X{\bf uu^T})(X-X{\bf uu^T})^T)\\
&=\underset{u \in R^d}{min}trace((X-X{\bf uu^T})(X^T-{\bf uu^T}X^T))\\
&=\underset{u \in R^d}{min}trace(XX^T-X{\bf uu^T}X^T-X{\bf uu^T}X^T+X{\bf uu^T}{\bf uu^T}X^T)\\
&=\underset{u \in R^d}{min}trace(XX^T-X{\bf uu^T}X^T)\\
&=\underset{u \in R^d}{min}trace(XX^T-{\bf u}^TX^TX{\bf u})\\
&=\underset{u \in R^d}{min}(trace(XX^T)-trace({\bf u}^TX^TX{\bf u}))\\
&=trace(XX^T)-\underset{u \in R^d}{max}trace({\bf u}^TX^TX{\bf u})\\
&=\underset{u \in R^d}{max}({\bf u}^TX^TX{\bf u})
\end{align}$$

Therefore, the minimization program and the maximization program are equivalent.
