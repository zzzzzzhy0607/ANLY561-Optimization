### 3.
The question $\underset{x\in{\mathbb{R}}}{max}=\frac{1}{2}{\bf x}^TA{\bf x}$ subject to $||x||^2=1$ is equivalent to the question
$\underset{x\in{\mathbb{R}}}{min}=-\frac{1}{2}{\bf x}^TA{\bf x}$ subject to $||x||^2=1$.

Assume 
$$
A=\begin{pmatrix}
a & b \\
b & d
\end{pmatrix}
,
{\bf x}=\begin{pmatrix}
x_1 \\
x_2
\end{pmatrix}
$$

Then $$f(x_1,x_2)=-\frac{1}{2}{\bf x}^TA{\bf x}=-\frac{1}{2}(ax_1^2+2bx_1x_2+dx_2^2)$$
$$f'(x_1,x_2)=-\frac{1}{2}\begin{pmatrix}
2a & 2b \\
2b & 2d
\end{pmatrix}
\begin{pmatrix}
x_1 \\
x_2
\end{pmatrix}
=-\begin{pmatrix}
a & b \\
b & d
\end{pmatrix}
\begin{pmatrix}
x_1 \\
x_2
\end{pmatrix}
=-A{\bf x}
$$

$f(x_1,x_2)$ subject to $||x||^2=1$, that is $f(x_1,x_2)$ subject to $g(x_1,x_2)=x_1^2+x_2^2-1$

Therefore $$g'(x_1,x_2)=2\begin{pmatrix}
x_1 \\
x_2
\end{pmatrix}$$

Let $$
-\begin{pmatrix}
a & b \\
b & d
\end{pmatrix}
\begin{pmatrix}
x_1 \\
x_2
\end{pmatrix}
=2\lambda\begin{pmatrix}
x_1 \\
x_2
\end{pmatrix}$$

$$-A{\bf x}=2\lambda{\bf x}$$

$$A{\bf x}=-2\lambda{\bf x}$$

Let $\alpha=-2\lambda$, then we have $A{\bf x}=\alpha{\bf x}$. Therefore $\alpha$ are the eigenvalues of A.

Assume we list all eigenvalues from largest to smallest: $\alpha_1, \alpha_2, ... \alpha_n$.

$$-\frac{1}{2}{\bf x}^TA{\bf x}=-\frac{1}{2}{\bf x}^T\alpha{\bf x}=-\frac{1}{2}\alpha{\bf x}^T{\bf x}=-\frac{1}{2}\alpha$$

When $-\frac{1}{2}{\bf x}^TA{\bf x}$ has the minimum value, $-\frac{1}{2}\alpha$ has the minimum value.

At this time, $\alpha=\alpha_1$, $ A{\bf x^*}=\alpha_1{\bf x^*}$, where ${\bf x^*}$ is the corresponding eigenvector of $\alpha_1$.

Therefore, the solution to this program is an eigenvector of A corresponding to the largest eigenvalue of A.
